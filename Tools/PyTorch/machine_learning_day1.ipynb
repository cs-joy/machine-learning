```python
# second most common errors in deep learning: shape error
```

```python
tensor_A = torch.tensor([[1,3],
                         [2,4],
                         [4, 5]])

tensor_B = torch.tensor([[2,4],
                         [4,6],
                         [6, 8]])
```

```python
torch.matmul(tensor_A, tensor_B)
```

```python
tensor_A.shape, tensor_B.shape # inner dimension don't match
```

```python
# let's solve that issue
tensor_B.T
```

```python
tensor_B.T.shape
```

```python
# now check the dimensions of two tensors
tensor_A.shape, tensor_B.T.shape
```

```python
# so now, it's ready to play
torch.matmul(tensor_A, tensor_B.T)
```

```python
# The matrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

```python
# let's transpose the tensor_A instead of tensor_B
torch.matmul(tensor_A.T, tensor_B)
```

```python
# reshaping, viewing and stacking, squeezing and unsqueezing tensors

# Reshaping  - reshapes an input tensor to defined shape
# View - return a view of an input tensor of certain shape but keep the memory as the original tensor
# Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
# Squeeze - removes all `1` dimensions from a tensor
# Unsqueeze - add a `1` dimension to a target tensor
# Permute - return a view of the input with dimensions permuted(swapped) in a certain way
```

```python
x = torch.arange(1., 10.)
x, x.shape
```

```python
# add an extra dimension
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(9,1)
x_reshaped, x_reshaped.shape
```

```python
# compatible
a = torch.arange(1., 13.)
print(a, a.shape)

a_reshaped = a.reshape(1, 12) # 1 x 12 = 12
print(a_reshaped, a_reshaped.shape)

a_reshaped = a.reshape(3, 4) # 3 x 4 = 12
print(a_reshaped, a_reshaped.shape)

a_reshaped = a.reshape(6, 2) # 6 x 2 = 12
print(a_reshaped, a_reshaped.shape)
```

```python
# change the view
z = x.view(1, 9)
z, z.shape
```

```python
# changing z changes x (because a view of a tensor shares the same memory as the original input)
z[: ,0] = 5
z, x
```

```python
# stack tensors on top of each other # https://pytorch.org/docs/stable/generated/torch.stack.html
x_stacked = torch.stack([x, x, x, x])
x_stacked
```

```python
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

```python
# squeeze
# remove 1 dimension from the original tensor
print(x_reshaped.shape)
x_squeezed = x_reshaped.squeeze()
x_squeezed.shape
```

```python
# unqueeze
# add single dimension
print(x_squeezed, x_squeezed.shape)
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(x_unsqueezed, x_unsqueezed.shape)
```

```python
# permute
# returns a view of the original tensor input with it's dimensions permuted

original_tensor = torch.rand(245, 223, 3) # [height, width, colour channels]
print(original_tensor.shape)
permuted_tensor = original_tensor.permute(2, 0, 1) # dimensions permuting: [2->0, 0->1, 1->2]
permuted_tensor.shape
```

```python
original_tensor[0, 0, 0] = 0.7254
original_tensor[0, 0 , 0], permuted_tensor[0, 0, 0]
```

```python
# PyTorch tensors and NumPy
  # data in numpy, want in PyTorch tensor -> torch.from_numpy(ndarray)
  # data in tensor, want in numoy -> torch.Tensor.numpy()

# warning: when converting from numpy -> pytorch, pytorch reflects numpy's default datatype of float64 unless specified otherwise

import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
# numpy default datatype is float64
array.dtype
```

```python
# tensor default datatype is float32
torch.arange(1.0, 8.0).dtype
```

```python
# change the value of numpy array, what will this do to `tensor`?
array = array + 1
array, tensor
```

```python
# tensor to numpy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# change the tensor, what happens to `numpy_tensor`?
tensor = tensor + 1
tensor, numpy_tensor
```
